```python
import numpy as np
```

```python
def find_sum_of_multiples(N, multiples):

  x = np.arange(1,N)

  idx_matrix = np.zeros((len(multiples), N-1))
  for idx in np.arange(len(multiples)):
    # idx_matrix[idx, :]
    idx_matrix[idx, :] = x % multiples[idx] == 0

  return np.dot(idx_matrix.max(axis=0), x).astype(int)
```

```python
find_sum_of_multiples(1000,[3,5])
```

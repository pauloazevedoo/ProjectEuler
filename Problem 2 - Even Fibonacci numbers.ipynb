```python
import numpy as np
```

```python
def find_sum_of_multiples(x, multiples):

  idx_matrix = np.zeros((len(multiples), len(x)))
  for idx in np.arange(len(multiples)):
    # idx_matrix[idx, :]
    idx_matrix[idx, :] = x % multiples[idx] == 0

  return np.dot(idx_matrix.max(axis=0), x).astype(int)
```

```python
def generate_fib_numbers(max, a, b):
  

  aux = []
  aux.append(a)
  aux.append(b)

  while aux[-1] <= max:
    aux.append(aux[-1]+aux[-2])


  return np.array(aux)
```

```python
x = generate_fib_numbers(4000000, 1, 2)
```

```python
find_sum_of_multiples(x, [2])
```
